```python
# выведем информацию о топ-10 категорий по просмотрам
top_category_view_df = df.groupby('category_id')[['views']].sum()\
    .sort_values('views', ascending=False) 
    
top_category_view_df.style.bar(align='mid')
```

```python
# создадим колонку вовлеченности пользователей: количество комментариев разделим на кол-во просмотров
# и посмотрим на категории 
#df['activity'] = (df['comment_count'] / df['views']) * 100

# наверное, более правильным активность пользователей суммировать из трех колонок: лайк, дизлайк и комментарии
df['activity'] = ((df['comment_count'] + df['likes'] + df['dislikes']) / df['views']) * 100
only_6_columns = df[['views', 'likes', 'dislikes', 'category_id', 'comment_count', 'activity']]
#df_new2 = df_new.groupby('category_id')[['activity']].sum() #.sort_values('activity', ascending=False)
#df_new2.plot(kind='bar')
#df_new.sort_values('activity', ascending=False).head(10).plot(kind='bar')
sum_in_category = only_6_columns.groupby('category_id')[['views', 'likes', 'dislikes', 'comment_count', 'activity']].sum()

# сгруппируем категории, суммируем и отсортируем по просмотрам
sum_in_category.sort_values('views', ascending=False).style.bar(align='mid')
```

```python
top_5_in_category = sum_in_category.sort_values('views', ascending=False) #.head(5)
top_5_in_category['sum_like_dislike'] = top_5_in_category['likes'] + top_5_in_category['dislikes']
top_5_in_category
```
